```python
def contiueNum(queue): 

    if len(queue) == 0:
        return 0
    packet = []
    turnNum = 0
    v = queue.pop(0)
    
    while len(queue)>0:
        vv = queue.pop(0)

        if v+1 == vv:
            v = vv
            turnNum = turnNum + 1
            if turnNum > 3:
                v = v - 4
                print(turnNum)
                packet.append(v)
        else:
            turnNum = 0
            v = vv
            
    return packet
```
