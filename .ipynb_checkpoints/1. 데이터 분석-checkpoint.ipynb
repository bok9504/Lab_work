```python
up_data_array = np.array(up_data)
```

```python
time_list = up_data.sort_values(by="통계시각")
time_list = time_list["통계시각"].unique()
sp_list = np.array([9,10,11,12])
```

```python
mean_spd_per_time_line1 = []
mean_spd_per_time_line2 = []
mean_spd_per_time_line3 = []
mean_spd_per_time_line4 = []

for i in sp_list:
    for it in time_list:
        data_for_time = up_data_array[(up_data_array[:,1] == it) & (up_data_array[:,2] == i), :]
        mean_spd = data_for_time[:,i].mean()
        mean_spd_per_time_line.append([it, mean_spd])
```

```python
s = pd.Series(mean_spd_per_time_line[:,1], index = mean_spd_per_time_line[:,0])
```
